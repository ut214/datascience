```
import pandas as pd
import numpy as np
np.random.seed(123)
n_dims= 10
n_samples = 1000
X_data = np.random.randn(n_samples,n_dims)
Tw = np.random.randn(n_dims,1)
Y_data = np.matmul(X_data,Tw)+np.random.randn(n_samples,1)
```

```
Y_data
```

```
x = tf.placeholder(tf.float32 , shape=(100,n_dims))
y = tf.placeholder(tf.float32,shape=(100,1))
W = tf.Variable(np.random.rand(10,1),dtype=tf.float32 )
temp  = tf.placeholder(tf.float32 ,shape=(100,n_dims))
```

```
Y_hat = tf.matmul(x,W)
Error = tf.reduce_sum((Y_hat-y)**2/n_samples)

prediction= tf.matmul(temp , W)
```

```
opt = tf.train.AdamOptimizer()

opt_operation = opt.minimize(Error)
```

```
with  tf.Session() as sess :
  sess.run(tf.initialize_all_variables())
  for i in range(10000):
    indices = np.random.choice(n_samples , 100)
    X_batch = X_data[indices]
    Y_batch = Y_data[indices]
    e,error = sess.run([opt_operation,Error],feed_dict={x : X_batch, y :Y_batch})

  predict_value = sess.run(prediction , feed_dict={temp:np.random.randn(100,10)})    
```
